```python
# Data Cleaning
corelation=airlines.corr()
```

```python
#Heatmap
sns.heatmap(corelation, xticklabels= corelation.columns, yticklabels=corelation.columns,
            annot=True)
```

```python
#Pairplot
sns.pairplot(airlines)
```

```python
#Relationship plot
sns.relplot(x='dep_time',y='sched_dep_time',hue='flight', data = airlines)
```

```python
# Test in Correlation
sns.regplot(x='sched_dep_time',y='dep_time',data=airlines)
```

```python
#Category plot
sns.catplot(x='arr_time',kind='box', data=airlines)
```
